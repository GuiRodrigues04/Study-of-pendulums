```python
# creating a list that represents 1000 values of the wire lenght that hold the mass(sphere) from the physics pendulum

D = 0
distancia_furo = []
for i in range(1000):
    D += 0.02
    distancia_furo.append(D)   
```

```python
# making the list T(periods) of the physics pendulum 

Periodos_fisico = []
g = 9.78
for i in distancia_furo:
    Periodos_fisico.append(2 * 3.14159 * ((i / g) ** (1 / 2)))
```

```python
# making the physics pendulum graph T(period) by position

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(distancia_furo, Periodos_fisico, color='C3', linestyle='--', linewidth=2 )
ax.grid(True)
plt.title("Traçado equação modelo do pêndulo físico")
plt.xlabel('Distância(D) em metros')
plt.ylabel('Periodo(T) em segundos')
```

```python
# creating a list that represents 1000 values of the wire lenght that hold the mass(sphere) from the simple pendulum

Periodos_simples = []
k = 0.2499
g = 9.78
for i in distancia_furo:
    Periodos_simples.append(2 * 3.14159 * (((i + ((k ** 2) / i)) / g) ** (1 / 2)))
```

```python
# making the simple pendulum graph T(period) by position

fig, aj = plt.subplots(figsize=(8,6))
aj.plot(distancia_furo, Periodos_simples, color='C4', linestyle='--', linewidth=2 )
aj.grid(True)
plt.title("Traçado equação modelo do pêndulo simples(comportamento inicial)")
plt.xlabel('Distância(D) em metros')
plt.ylabel('Periodo(T) em segundos')
```

```python
# comparing both graphs

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(distancia_furo, Periodos_simples, color='C2', linewidth= 2, label='Pendulo simples')
ax.plot(distancia_furo, Periodos_fisico, color='C6', linestyle='--', linewidth=2, label="Pêndulo Físico")
ax.grid(True)
plt.title("Comparação entre pêndulos com valores de 'g' iguais")
plt.xlabel('Distância(D) em metros')
plt.ylabel('Periodo(T) em segundos')
ax.legend()
```
